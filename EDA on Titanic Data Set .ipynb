# Survival with respect to Pclass

```python
sns.countplot(x="Pclass",hue="Survived",data=df,palette="icefire",color="b")
plt.show()
```

# Survival with respect to Gender

```python
sns.countplot(x="Sex",hue="Survived",data=df,palette="Blues",color="g")
plt.show()
```

# Survival with respect to Age

```python
sns.distplot(df[df["Survived"]==0]["Age"],color="r")
sns.distplot(df[df["Survived"]==1]["Age"],color="b")
plt.show()
```
